```python
# Merging Columns
```

```python
df_ords_prods_merge["user_id"] = df_ords_prods_merge["user_id_x"].fillna(df_ords_prods_merge["user_id_y"])
df_ords_prods_merge = df_ords_prods_merge.drop(columns=["user_id_x", "user_id_y"])
```

```python
df_ords_prods_merge.columns
```

```python
## Making sure both columns have the same data type
```

```python
print(type(df_ords_prods_merge))
print(type(df_custs))
```

```python
print(df_ords_prods_merge.columns.tolist())
print(df_custs.columns.tolist())
```

```python
if isinstance(df_ords_prods_merge, pl.DataFrame):
    df_ords_prods_merge = df_ords_prods_merge.to_pandas()

if isinstance(df_custs, pl.DataFrame):
    df_custs = df_custs.to_pandas()

print(type(df_ords_prods_merge))
print(type(df_custs))
```

```python
user_col_merge = [c for c in df_ords_prods_merge.columns if "user_id" in c][0]

user_col_cust = [c for c in df_custs.columns if "user_id" in c][0]

print("Orders/Products user column =", user_col_merge)
print("Customers user column      =", user_col_cust)
```

```python
df_ords_prods_merge["user_id"] = df_ords_prods_merge["user_id"].astype("int64")
df_custs["user_id"] = df_custs["user_id"].astype("int64")

df_ords_prods_merge_custs = df_ords_prods_merge.merge(
    df_custs,
    on="user_id",
    how="inner"
)
```
